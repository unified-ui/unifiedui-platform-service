# Vault Client Test

Test HashiCorp Vault client initialization and secret operations

```python
from unifiedui.vault.hashicorp_vault.client import HashiCorpVaultClient

vault_client = HashiCorpVaultClient(
    url="http://localhost:8200",
    token="admin",
    mount_point="secret"
)

print("Vault client initialized successfully")
```

```python
test_key = "test/tenant-123/credential-456"
test_secret = "my-super-secret-api-key-12345"
test_metadata = {"description": "Test credential", "type": "API_KEY"}

vault_uri = vault_client.store_secret(
    key=test_key,
    value=test_secret,
    metadata=test_metadata
)

print(f"Secret stored successfully!")
print(f"Vault URI: {vault_uri}")
```

```python
retrieved_secret = vault_client.get_secret(vault_uri)

print(f"Secret retrieved successfully!")
print(f"Secret value: {retrieved_secret}")
print(f"Matches original: {retrieved_secret == test_secret}")
```
